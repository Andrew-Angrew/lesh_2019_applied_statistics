```python
from scipy.stats import bernoulli
import numpy as np

from matplotlib import pyplot as plt
```

```python
distr = bernoulli(0.3)
sample_size = 1000
sample = distr.rvs(sample_size)
prefix_means = np.cumsum(sample) / np.arange(1, sample_size + 1)

plt.rc("figure", figsize=(16, 10))
plt.xlabel("Размер выборки")
plt.plot(prefix_means, label="Выборочное среднее")
plt.hlines(distr.mean(), 0, sample_size, label="Истиное среднее")
plt.hlines(0, 0, sample_size, color="white")
plt.hlines(1, 0, sample_size, color="white")
plt.legend()
plt.show()
```

```python
is_diffent = bernoulli(0.5).rvs()
distr1 = bernoulli(0.5)
distr2 = bernoulli(0.4) if is_diffent else distr1
if bernoulli(0.5).rvs() > 0:
    distr1, distr2 = distr2, distr1

sample_size = 1000

sample1 = distr1.rvs(sample_size)
sample2 = distr2.rvs(sample_size)
prefix_means1 = np.cumsum(sample1) / np.arange(1, sample_size + 1)
prefix_means2 = np.cumsum(sample2) / np.arange(1, sample_size + 1)
```

```python
show_size = 100
show_size = sample_size

plt.rc("figure", figsize=(16, 10))
plt.xlabel("Размер выборки")
plt.plot(prefix_means1[:show_size], label="Средние из первой выборки")
plt.plot(prefix_means2[:show_size], label="Средние из второй выборки")
# plt.hlines(distr1.mean(), 0, show_size, label="Cреднее первой выборки")
# plt.hlines(distr2.mean(), 0, show_size, label="Cреднее второй выборки")
plt.hlines(0, 0, show_size, color="white")
plt.hlines(1, 0, show_size, color="white")
plt.legend()
plt.show()
```
